```python
#import Natural Lanaguage Toolkit
import nltk
```

```python
#tokenize
nltk.word_tokenize(data[1][1])
```

```python
#tokenize with whitespace
nltk.tokenize.WhitespaceTokenizer().tokenize(data[1][1])
```

```python
# tokenize using Punctuation
nltk.tokenize.WordPunctTokenizer().tokenize(data[1][1])
```

```python
# tokenize using grammar rules
nltk.tokenize.TreebankWordTokenizer().tokenize(data[1][1])
```

Token normalization with stemming and lemmatization


```python
words = nltk.tokenize.WhitespaceTokenizer().tokenize(data[1][1])
```

```python
df = pd.DataFrame()
df['Original Words'] = pd.Series(words)
```

```python
porterStemmedWords = [nltk.stem.PorterStemmer().stem(word) for word in words]
```

```python
df['PorterStemmedWords'] = pd.Series(porterStemmedWords)
```

```python
df['PorterStemmedWords']
```

```python
#WordNet Lemmatization
wordNetLemmatizedWords = [nltk.stem.WordNetLemmatizer().lemmatize(word) for word in words]
```

```python
df['WordNetLemmatizer'] = pd.Series(wordNetLemmatizedWords)
df
```
